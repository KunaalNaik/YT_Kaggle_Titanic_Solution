```python
import pandas as pd
```

```python
train = pd.read_csv("excel_full_train.csv")
train.head(1)
```

```python
test = pd.read_csv("excel_test.csv")
test.head(1)
```

```python
X = train.drop(['PassengerId','Survived'], axis = 1)
```

```python
y = train['Survived']
```

```python
#### Use Test Train Split to divide into train and test
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=21)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treeModel = DecisionTreeClassifier()
```

```python
treeModel.fit(X_train, y_train)
```

### Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'criterion' : ('gini', 'entropy'), 'max_depth' : (3,4,5,6,7,8,9,10),
              'min_samples_leaf' : (1,2,3,4,5)}
```

```python
DT_grid  = GridSearchCV(DecisionTreeClassifier(), param_grid = parameters, cv = 3)
```

```python
DT_grid_model = DT_grid.fit(X_train, y_train)
```

```python
DT_grid_model.best_estimator_
```

```python
DT_grid_model.best_score_
```

```python
treeModel = DecisionTreeClassifier(class_weight=None, criterion='gini', max_depth=3,
                       max_features=None, max_leaf_nodes=None,
                       min_impurity_decrease=0.0, min_impurity_split=None,
                       min_samples_leaf=1, min_samples_split=2,
                       min_weight_fraction_leaf=0.0, presort=False,
                       random_state=None, splitter='best')
```

```python
treeModel.fit(X_train, y_train)
```

```python
treeModel.score(X_train, y_train)
```

```python
best_feat = pd.DataFrame({'Features': X_train.columns,'Importance': treeModel.feature_importances_})
best_feat.sort_values('Importance', ascending = False).head(10)
```

### Evaluation of Test

```python
X_test.head(1)
```

```python
y_pred = treeModel.predict(X_test)
```

```python
#from sklearn.metrics import classification_report
#print(classification_report(y_test,y_pred ))
```

```python
print(f'Test : {treeModel.score(X_test, y_test):.3f}')
print(f'Train : {treeModel.score(X_train, y_train):.3f}')
```

### Prediction of Test

```python
sub_test = test.drop(['PassengerId'], axis = 1)
```

```python
sub_test_pred = treeModel.predict(sub_test).astype(int)
```

```python
AllSub = pd.DataFrame({ 'PassengerId': test['PassengerId'],
                       'Survived' : sub_test_pred
    
})

AllSub.to_csv("All_Var_Video_DT_grid.csv", index = False)
```
